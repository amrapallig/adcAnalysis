```python
#density
def density_eos(t,s):
    alpha=2e-4; beta=8e-4; g=9.81; rho=1026.0
    density= rho*(1.0-alpha*(t-20)+beta*(s-35))
    return density

# buoyancy flux
def wb_eos(wt_sfc,ws_sfc):
    alpha=2e-4; beta=8e-4; g=9.81; rho=1026.0
    wb_sfc=g*(alpha*wt_sfc-beta*ws_sfc)
    #wb=g*(alpha*wt-beta*ws)
    return wb_sfc

#OSBL depth
def bld(t,s,z):
#     ind=np.where((z)>=-70)
#     s=s[ind]; t=t[ind]; z=z[ind]
    density=density_eos(t,s)
    #print(z.shape)
    N2=np.gradient(-density,z,axis=1)
    bld=np.zeros(len(N2[:,0]))
    for t in np.arange(len(N2[:,0])):    
        indx=N2[t,:].argmax()
        bld[t]=z[indx]
    #bld2=np.where(bld>50,nan,bld)
        #print(bld.shape)
    return np.nanmean(bld)  # depth corresponding to max N2 for all time t

#wstar
def wstar_sfc(wb,h):
    ws=(wb*(-h))**(1/3)
    #wb=g*(alpha*wt-beta*ws)
    return np.nanmean(ws**2)
```

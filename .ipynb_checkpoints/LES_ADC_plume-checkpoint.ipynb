```python
import matplotlib.pyplot as plt # plotting
import numpy as np              # working with number arrays etc.
from netCDF4 import Dataset     # reading netcdf files
import pandas as pd  
import seaborn as sns

# path for data
path_les= "/Users/amrapallig/Desktop/cvmix/DATA_ADC_paper1/LES/"
path_adc= "/Users/amrapallig/Desktop/cvmix/DATA_ADC_paper1/ADC/ADC_allRez/"
```

```python
#w3 is at cell center and other second-moment fluxes at cell interface
# 
def plume_les(w3,w2,wt,ws,zm):
    w3avg=0.5*(w3[1:]+w3[:-1])
    sw=w3avg/(w2[1:]**(3/2))
    sigma=0.5+sw/(2*np.sqrt(4+sw**2))
    wumd=np.sqrt(w2[1:]/(sigma*(1-sigma)))
    wumd=np.sqrt(w2[1:]/(4+sw**2))/(sigma*(1-sigma))
    
    mc=sigma*(1-sigma)*wumd
    tumd=wt[1:]/mc
    sumd=ws[1:]/mc
    
    return sigma,mc,wumd, tumd, sumd, sw, zm[1:]
  
```

```python
t=96

les= Dataset(path_les+"t1s15.nc", mode='r')
fh=Dataset(path_adc+'output.0001-01-01_00.00_1m.nc', mode='r')
mf= Dataset(path_adc+'adcTurbulenceArrays_t1s15_1m.nc', mode='r')  

#zm_les=les.variables['zt'][:]
sigma_les, mc_les,wumd_les, tumd_les, sumd_les,sw_les,zm_les=plume_les(les.variables['w3'][t,:], les.variables['w2'][t,:],les.variables['wt'][t,:],les.variables['ws'][t,:],les.variables['zm'][:])

## ADC
sigma_adc=mf.variables["areaFraction"][t,0,:]; mc_adc=mf.variables["Mc"][t,0,:];wumd_adc=mf.variables["wumd"][t,0,:];
tumd_adc=mf.variables["tumd"][t,0,:]; sumd_adc=mf.variables["sumd"][t,0,:];zm_adc=fh.variables["zTop"][t,0,:];

wt_sps=sigma_adc*mf.variables["wt_spsD"][t,0,:]+(1-sigma_adc)*mf.variables["wt_spsU"][t,0,:]
ws_sps=sigma_adc*mf.variables["ws_spsD"][t,0,:]+(1-sigma_adc)*mf.variables["ws_spsU"][t,0,:]

tumd_adc_sps=wt_sps/mc_adc
sumd_adc_sps=ws_sps/mc_adc
```
